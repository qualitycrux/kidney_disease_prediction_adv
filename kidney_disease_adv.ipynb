```python
#kidney disease prediction application
#import tools
import numpy as np
import pandas as pd

#preprocesing tools
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split

#classification models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier

#for model evaluation
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
#load data
df=pd.read_csv("/content/drive/MyDrive/dataset/kidney_disease.csv")

filtered_df = df[df["classification"] != "ckd"]

df.head(100)
```

```python
important_columns=['age','bp','sg','al','hemo','pcv','rc','bu','bgr','sc','sod','pot','htn','dm','cad','appet','pc','pe','ane','classification']
df=df[important_columns]
df.head(200)
```

```python
# data preprocessing
# 1. remove \t from all values
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip().str.replace('\t', '', regex=True)

```

```python
#2. check isnull
df.isnull().sum()
```

```python
import numpy as np
#3  fill missing values
# Convert 'pcv' and 'rc' to numeric, coercing errors to NaN
df['pcv'] = pd.to_numeric(df['pcv'], errors='coerce')
df['rc'] = pd.to_numeric(df['rc'], errors='coerce')

df['age'] = df['age'].fillna(df['age'].median())
df['bp'] = df['bp'].fillna(df['bp'].median())
df['sg'] = df['sg'].fillna(df['sg'].median())
df['al'] = df['al'].fillna(df['al'].median())
df['hemo'] = df['hemo'].fillna(df['hemo'].median())
df['pcv'] = df['pcv'].fillna(df['pcv'].median())
df['rc'] = df['rc'].fillna(df['rc'].median())
df['bu'] = df['bu'].fillna(df['bu'].median())
df['bgr'] = df['bgr'].fillna(df['bgr'].median())
df['sc'] = df['sc'].fillna(df['sc'].median())
df['sod'] = df['sod'].fillna(df['sod'].median())
df['pot'] = df['pot'].fillna(df['pot'].median())
df['htn'] = df['htn'].fillna(df['htn'].mode()[0])
df['dm'] = df['dm'].fillna(df['dm'].mode()[0])
df['cad'] = df['cad'].fillna(df['cad'].mode()[0])
df['appet'] = df['appet'].fillna(df['appet'].mode()[0])
df['pc'] = df['pc'].fillna(df['pc'].mode()[0])
df['pe'] = df['pe'].fillna(df['pe'].mode()[0])
df['ane'] = df['ane'].fillna(df['ane'].mode()[0])
df['classification'] = df['classification'].fillna(df['classification'].mode()[0])
df.isnull().sum()
#df.head()
```

```python
# lable encoding categorical values
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['htn'] = le.fit_transform(df['htn'])
df['dm'] = le.fit_transform(df['dm'])
df['cad'] = le.fit_transform(df['cad'])
df['classification'] = le.fit_transform(df['classification'])
df['appet'] = le.fit_transform(df['appet'])
df['pc'] = le.fit_transform(df['pc'])
df['pe'] = le.fit_transform(df['pe'])
df['ane'] = le.fit_transform(df['ane'])
df.head(200)



```

```python
# scalling the data
numerical_columns=['age','bp','sg','al','hemo','pcv','rc','bu','bgr','sc','sod','pot']

scaler=MinMaxScaler()
df[numerical_columns]=scaler.fit_transform(df[numerical_columns])
df
```

```python
#balance the data for target value
df['classification'].value_counts()
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=42)
X=df.drop('classification',axis=1)
y=df['classification']

X_balanced,y_balanced=smote.fit_resample(X,y)
y_balanced.value_counts()
```

```python
# split test and train
X_train,X_test,y_train,y_test=train_test_split(X_balanced,y_balanced,test_size=0.2,random_state=42)

#check the shape
print(X_train.shape,X_test.shape)
```

```python
# train multiple classfication models
models = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Support Vector Machine': SVC(),
    'Naive Bayes': GaussianNB(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'AdaBoost': AdaBoostClassifier()
}

# train and evaluate each model
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    class_report = classification_report(y_test, y_pred)

    # Print the results
    print(f"Model: {name}")
    print(f"Accuracy: {accuracy}")
    print("Confusion Matrix:")
    print(conf_matrix)
    print("Classification Report:")
    print(class_report)
    print("="*50)

```

```python
#selecting best model

model_gbc= GradientBoostingClassifier()
model_gbc.fit(X_train,y_train)
y_pred=model_gbc.predict(X_test)
print(confusion_matrix(y_test,y_pred))
```

```python

#load the model and scaller
model=pickle.load(open('/content/drive/MyDrive/Models/model-kidney-disease.pkl','rb'))
scaller=pickle.load(open('/content/drive/MyDrive/Models/scaller-kidney-disease.pkl','rb'))

#Prediction on new data using function
#def predict_kidney_disease(age, bp, sg, al, hemo, sc,sod,pot, htn, dm, cad, appet, pc,pe):
def predict_kidney_disease(age, bp, sg, al, hemo,pcv,rc,bu,bgr, sc,sod,pot, htn, dm, cad, appet, pc,pe,ane):
    # Create a DataFrame with input variables, following the correct order
    df_dict = {
        'age': [age],
        'bp': [bp],
        'sg': [sg],
        'al': [al],
        'hemo': [hemo],
        'pcv': [pcv],
        'rc': [rc],
        'bu': [bu],
        'bgr': [bgr],
        'sc': [sc],
        'sod':[sod],
        'pot':[pot],
        'htn': [htn],
        'dm': [dm],
        'cad': [cad],
        'appet': [appet],
        'pc': [pc],
        'pe' : [pe],
        'ane': [ane]
    }
    df = pd.DataFrame(df_dict)

    # Encode the categorical columns
    le = LabelEncoder()
    df['htn'] = le.fit_transform(df['htn'])
    df['dm'] = le.fit_transform(df['dm'])
    df['cad'] = le.fit_transform(df['cad'])
    df['appet'] = le.fit_transform(df['appet'])
    df['pc'] = le.fit_transform(df['pc'])
    df['pe'] = le.fit_transform(df['pe'])
    df['ane'] =le.fit_transform(df['ane'])

    # Scale the numeric columns using the previously fitted scaler
    numeric_cols=['age','bp','sg','al','hemo','pcv','rc','bu','bgr','sc','sod','pot']
    df[numeric_cols] = scaller.transform(df[numeric_cols])

    # Make the prediction
    prediction = model.predict(df)

    # Return the predicted class
    return prediction[0]
```

```python
#predict the results now giving varable values
#ckd sample data:
#   62.0	80.0	1.010	2.0	9.6	    1.8 138.0	4.4	no	yes	no	poor	normal	    ckd
#	48.0	70.0	1.005	4.0	11.2	3.8	138.0	4.4 yes	no	no	poor	abnormal	ckd
#---------No CKD----------
#	55.0	80.0	1.020	0.0	15.7	0.5	no	no	no	good	normal	notckd
#	42.0	70.0	1.025	0.0	16.5	1.2	no	no	no	good	normal	notckd
input_data = {
    'age': 25,          # 1.Age in years (older age is associated with a higher risk of CKD)
    'bp': 80,           # 2.Blood pressure in mmHg (high blood pressure is a major risk factor)
    'sg': 1.020,        # 3.Urine specific gravity (low or abnormal values may indicate impaired kidney function)
    'al': 0.0,          # 4.Urine albumin level (protein in urine; elevated levels suggest kidney damage)
    'hemo': 15.4,       # 5.Hemoglobin level in g/dL (low levels may indicate anemia, common in CKD)
    'pcv'  :44.0,       # 6.Packed Cell Volume (PCV)
    'rc' : 5.2,         # 7.Red Blood Cell Count (RBC)
    'bu': 36.0,         # 8.Blood Urea
    'bgr': 121.0,       # 9.Random Blood Glucose
    'sc': 0.5,          # 10.Serum creatinine in mg/dL (higher values indicate reduced kidney function)
    'sod':100.0,        # 11. Sodium
    'pot':4.5,          # 12. Potassium
    'htn': 'no',        # 13.Hypertension ('yes' if the patient has high blood pressure)
    'dm': 'no',         # 14.Diabetes mellitus ('yes' if the patient has diabetes, a major CKD risk factor)
    'cad': 'no',        # 15.Coronary artery disease ('yes' if the patient has heart disease)
    'appet': 'good',    # 16.Appetite ('good' or 'poor'; poor appetite is common in advanced CKD)
    'pc': 'normal',     # 17.Pus cells in urine ('normal' or 'abnormal'; abnormal may indicate infection or inflammation)
    'pe': 'no ',        # 18.Yes may indicate kidney desease
    'ane': 'no',        # 19. aanemia
}
result = predict_kidney_disease(**input_data)
if result == 0: # 0 represents 'ckd' based on label encoding
    print("Patient has kidney disease")
else:
  print("Patient does not have kidney disease")
```
